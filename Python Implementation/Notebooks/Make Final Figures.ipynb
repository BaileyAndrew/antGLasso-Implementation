# Runtime Figures

```python
ahs3ds = np.load(
    f'../../Data for Plots/RUN ant hard small 3D sizes.npy'
)
ths3ds = np.load(
    f'../../Data for Plots/RUN tera hard small 3D sizes.npy'
)
aes3ds = np.load(
    f'../../Data for Plots/RUN ant easy small 3D sizes.npy'
)
tes3ds = np.load(
    f'../../Data for Plots/RUN tera easy small 3D sizes.npy'
)
ahs3d = np.load(
    f'../../Data for Plots/RUN ant hard small 3D.npy'
)
ths3d = np.load(
    f'../../Data for Plots/RUN tera hard small 3D.npy'
)
aes3d = np.load(
    f'../../Data for Plots/RUN ant easy small 3D.npy'
)
tes3d = np.load(
    f'../../Data for Plots/RUN tera easy small 3D.npy'
)
```

```python
with plt.style.context('Solarize_Light2'):
        plt.rcParams['axes.prop_cycle'] = colorscheme
        fig1, (ax1, ax2) = plt.subplots(figsize=(16, 8), ncols=2)
        ax1.plot(
            ahs3ds,
            ahs3d,
            label='antGLasso'
        )
        ax1.plot(ths3ds,
                 ths3d,
                 label='TeraLasso'
        )
        ax1.set_xlabel("Size of Precision Matrices")
        ax1.set_ylabel("Seconds")
        ax1.set_title(f"Runtime for small-sample [n*n*n] hard data")
        ax1.set_ylim([0, 60])
        ax1.legend()
        
        ax2.plot(
            aes3ds,
            aes3d,
            label='antGLasso'
        )
        ax2.plot(tes3ds,
                 tes3d,
                 label='TeraLasso'
        )
        ax2.set_xlabel("Size of Precision Matrices")
        ax2.set_ylabel("Seconds")
        ax2.set_title(f"Runtime for small-sample [n*n*n] easy data")
        ax2.set_ylim([0, 60])
        ax2.legend()
        
        fig1.savefig(f'../../Plots/Final/All Together Tensors')
```
